```python
import numpy as np
import math as mt
import scipy
import pandas
import matplotlib.pyplot as plt
from scipy.optimize import golden, minimize
%matplotlib inline
```

```python
def f(x):
    f = mt.sin(x/5) * mt.exp(x/10) + 5*mt.exp(-x/2)
    return(f)

def h(x):
    return int(f(x))
```

```python
res = minimize(h, 30, method='BFGS')
bounds = [(1,30)]
res_dif = scipy.optimize.differential_evolution(h, bounds)
```

```python
x = np.arange(1,30,1)
h2 = np.vectorize(h)
plt.plot(x, h2(x)) #Построение графика
plt.ylim([-12, 5])
plt.xlabel(r'$x$') #Метка по оси x в формате TeX
plt.ylabel(r'$f(x)$') #Метка по оси y в формате TeX
plt.grid(True) #Сетка
plt.show() #Показать график
```
